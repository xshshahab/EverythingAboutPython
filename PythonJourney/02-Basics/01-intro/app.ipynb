# üêç Python Data Types

In Python, **data types** specify the kind of value a variable holds. Python is **dynamically typed**, meaning you don't need to declare a variable's type ‚Äî Python figures it out at runtime.

---

## üì¶ Built-in Data Types

### 1. Numeric Types

- `int` ‚Äì Integer numbers  
- `float` ‚Äì Decimal numbers  
- `complex` ‚Äì Complex numbers

```python
x = 10        # int
y = 3.14      # float
z = 2 + 3j    # complex
```

---

### 2. Text Type

- `str` ‚Äì String (sequence of characters)

```python
name = "Python is awesome"
```

---

### 3. Boolean Type

- `bool` ‚Äì Boolean values: `True` or `False`

```python
is_active = True
is_admin = False
```

---

### 4. Sequence Types

- `list` ‚Äì Ordered, mutable sequence  
- `tuple` ‚Äì Ordered, immutable sequence  
- `range` ‚Äì Immutable sequence of numbers

```python
fruits = ["apple", "banana", "cherry"]  # list
coordinates = (10, 20)                  # tuple
numbers = range(5)                      # range(0, 5)
```

---

### 5. Mapping Type

- `dict` ‚Äì Collection of key-value pairs

```python
person = {
    "name": "Alice",
    "age": 30
}
```

---

### 6. Set Types

- `set` ‚Äì Unordered, unique items  
- `frozenset` ‚Äì Immutable set

```python
colors = {"red", "green", "blue"}        # set
frozen_colors = frozenset(["red", "green"])
```

---

### 7. None Type

- `NoneType` ‚Äì Represents the absence of a value

```python
result = None
```

---

## üîç Check Data Type

Use `type()` to check the data type of any variable:

```python
print(type(10))         # <class 'int'>
print(type("hello"))    # <class 'str'>
print(type([1, 2, 3]))   # <class 'list'>
```
