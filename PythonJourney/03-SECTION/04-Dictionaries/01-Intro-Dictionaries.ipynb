A **dictionary** in Python is a built-in data structure that stores data in **key-value pairs**. It's one of the most powerful and flexible container types in Python.

---

### âœ… **1. What is a Dictionary?**

A dictionary is like a real dictionary â€” it maps a **word (key)** to its **definition (value)**.

#### ðŸ”¹ Syntax:

```python
my_dict = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}
```

* Keys: `"name"`, `"age"`, `"city"` (must be unique and immutable)
* Values: `"Alice"`, `25`, `"New York"` (can be any type)

---

### âœ… **2. Creating a Dictionary**

```python
# Empty dictionary
empty_dict = {}

# Dictionary with data
person = {
    "name": "John",
    "age": 30,
    "is_student": False
}
```

---

### âœ… **3. Accessing Values**

```python
print(person["name"])       # Output: John
print(person.get("age"))    # Output: 30
```

* `[]` will give an error if the key doesnâ€™t exist.
* `.get()` will return `None` or a default value instead of an error.

---

### âœ… **4. Modifying Values**

```python
person["age"] = 31
```

---

### âœ… **5. Adding New Key-Value Pairs**

```python
person["gender"] = "Male"
```

---

### âœ… **6. Removing Items**

```python
# Remove a key
del person["is_student"]

# Using pop()
age = person.pop("age")
```

---

### âœ… **7. Dictionary Methods**

| Method               | Description                          |
| -------------------- | ------------------------------------ |
| `dict.keys()`        | Returns a list of all keys           |
| `dict.values()`      | Returns a list of all values         |
| `dict.items()`       | Returns a list of key-value pairs    |
| `dict.get(key)`      | Gets the value for a key             |
| `dict.pop(key)`      | Removes a key and returns its value  |
| `dict.update({...})` | Updates the dictionary with new data |

---

### âœ… **8. Looping Through a Dictionary**

```python
for key in person:
    print(key, person[key])

# OR
for key, value in person.items():
    print(f"{key}: {value}")
```

---

### âœ… **9. Nested Dictionaries**

```python
students = {
    "student1": {"name": "Alice", "age": 20},
    "student2": {"name": "Bob", "age": 22}
}
```

---

### âœ… **10. Dictionary Comprehension**

```python
squares = {x: x*x for x in range(1, 6)}
print(squares)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```
